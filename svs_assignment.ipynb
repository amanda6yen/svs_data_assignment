```python
import random

def generate_date(month, year):
   

    if month in [1,3,5,7,8,10,12]:
        day = random.randint(1, 31)
    elif month == 2:
        if year%4 == 0:
            if year%100 == 0:
                if year%400 == 0:
                    day = random.randint(1,29)
                else:
                    day = random.randint(1,28)
            day = random.randint(1,29)
        else:
            day = random.randint(1,28)
    else:
        day = random.randint(1,30)
   
    if len(str(month)) == 1:
        str_month = '0' + str(month)
    else:
        str_month = str(month)

    if len(str(day)) == 1:
        str_day = '0' + str(day)
    else:
        str_day = str(day)

    return_string = str_month + '/' + str_day + '/' + str(year)
    return return_string

#1.2 starts here

def check_match(dates):
    
#     Parameters:
#         dates: list of dates in the format 'mm/dd/yyyy'
#     Returns:
#         a tuple containing the number of matching birthdays in the original list
#         and a list of all the birthdays that match
    
    num_matches = 0
    matching_bdays = []
    for n in range(len(dates)):
        date = dates[n]
        birthdate = date[0:5]
        for i in range(n+1, len(dates)):
            if birthdate == dates[i][0:5]:
                if birthdate not in matching_bdays:
                    num_matches += 1
                    matching_bdays.append(birthdate)
    return (num_matches, matching_bdays)


#1.3 starts here

trials = int(input("How many trials do you want to run? "))
bdays = int(input("How many birthdays do you want to generate on each trial? "))
num_matches = 0

for n in range(trials):
    dates = []
    for i in range(bdays):
        m = random.randint(1,12)
        y = random.randint(1903, 2019) 
        dates.append(generate_date(m, y))
    sol = check_match(dates)
    if sol[0] >= 1:
        num_matches += 1
    print("Matches: " + str(sol[0]))
    
    clean_list = str(sol[1])
    clean_list = clean_list[1:len(clean_list)-1]
    print("Matching birthdays: " + clean_list)
    

percentage = num_matches/trials
print("Percentage of trials with at least one match: " + str(percentage*100) + "%")
```
